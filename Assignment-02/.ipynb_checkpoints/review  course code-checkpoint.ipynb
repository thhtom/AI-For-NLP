```python
import re
```

```python
def token(string):
    return ' '.join(re.findall('[\w|\d]+',string))
```

```python
all_articles = [token(str(a)) for a in all_articles]
```

```python
all_articles[:5]
```

```python
text = ''

for a in all_articles:
    text += a 
print ('length of text:{}'.format(len(text)))
```

```python
TEXT = text
```
