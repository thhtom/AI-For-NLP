```python
result = nlp.annotate(text, properties={ 'annotators': 'ner, pos',
                       'outputFormat': 'json',
                       'timeout': 1000,})
```

```python
result['sentences']
```

```python
pos_ner = []
for word in result['sentences'][0]['tokens']:
    pos_ner.append('{} ({})'.format(word['word'],word['ner']))
' '.join(pos_ner)
```

```python
pos = []
for word in result['sentences'][1]['tokens']:
    pos.append('{} ({})'.format(word['word'],word['pos']))
' '.join(pos)
    
```
